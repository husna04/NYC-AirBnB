# Menghilangkan(drop) baris yang memiliki data kosong (NaN)

*   buat copy nya
*   lalu hapus baris yang memiliki data kosong, di file yang sudah di copy



```python
dropped_df = df.copy()
#axis=0 untuk drop baris, axis=1 untuk drop kolom
dropped_df.dropna(axis=0, inplace=True)
dropped_df 
```

```python
dropped_df.info()
```

# Mengubah kolom price (string) ke dalam bentuk float


```python
dropped_df['price'].sum()
```

```python
dropped_df['prices']=dropped_df['price'].str.strip(' dollars')
dropped_df['prices'].sum()
```

```python
dropped_df['prices']=dropped_df['prices'].astype('float')
dropped_df.dtypes
```

```python
dropped_df['prices'].sum()
```

```python
df2 = dropped_df.drop(columns = 'price')
```

```python
df2.head()
```

# Mengecek waktu pada kolom last_review

```python
df2.head()
```

```python
df2.dtypes
```

last_review seharusnya merupakan tipe data datetime (tanggal+waktu) bukan object(string)

---

konversikan last_review dari string menjadi tipe data datetime dengan menggunakan `to_datetime` dan ubah menjadi date (tanpa waktu) dengan menambahkan `.dt.date` di akhir

> menambahkan `.dt.date` akan membuat tipe data menjadi object (string)



```python
df2['last_review'] = pd.to_datetime(df2['last_review']).dt.date
#df['last_review'] = pd.DatetimeIndex(df['last_review']).year
df2.dtypes
df2.head()
```

```python
df2.dtypes
```

Batas tanggal pada kolom last_review adalah hari ini (per dilakukan analisis 2022-05-14), apabila melebihi tanggal tersebut akan ada tindakan lebih lanjut.

```python
import datetime as dt
today_date = dt.date.today()
df2[df2['last_review'] > today_date]
```

Dalam perusahaan hal seperti ini perlu dilakukan konfirmasi terlebih dahulu, apakah ada salah input? gangguan teknis kah? dsb.

---
Dalam kasus ini, yang akan dilakukan adalah mengubah tanggal yang melebihi hari ini menjadi tanggal hari ini.


```python
df2.loc[df2['last_review']>today_date, 'last_review']=today_date
```

```python
df2[df2['last_review']>today_date]
```
